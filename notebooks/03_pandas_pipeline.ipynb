## 3 Ingeniería de datos con pandas

```python
# Importar las libreria necesarias que vamos a utilizar
import numpy as np
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy import text
```

#### **3.1 Carga el CSV con read_csv*

```python
# Cargamos el DataSet

df = pd.read_csv("../data/ventas_techventas.csv", sep=',', encoding='utf-8', parse_dates=['fecha'])
df.head()
```

```python
# Muestra tipo de datos, cuantos no nulos hay, y el uso de memoria
# Se muestra que la unica columna que no hay nulos es order_id, las demas columnas tienen 11 nulos cada una
df.info()
```

#### **3.2 Dectectar, eliminar e imputar nulos*

```python
# Utilizamoe la función isna() para conocer lo valores nulos en cada columna.
# Si es False significa no nulo, si es true hay nulo
df.isna()
```

```python
# Agregando la función sum() lo que me permite es contar cuantos nulos hay en cada columna
df.isna().sum()
```

```python
#Muestra que filas exactamente tiene nulos, en este caso estamos revisando con la columna cantidad
df[df['cantidad'].isna()]
```

```python
#Regla de oro: Nunca elimines un nulo sin enteder por qué está ahí!!!!!!!!!!!

#Buena practica: Siempre trabaja sobre una copia para no dañar el archivo orginal

#Creación de copias.
df_1 = df.copy()
```

```python
#Nos permite analizar los valores estadisticos descriptivos de los datos
df_1.describe()
```

```python
# Contar registros iniciales para el reporte
registros_iniciales = len(df_1)

# 1. Imputar la fila 42 por el motivo de que en producto tiene valor nulo y según la observación el unico prodcuto de categoria Electronica es "Lapton Pro 15"
fila_especifica = 42
df_1.loc[fila_especifica, 'producto'] = 'Laptop Pro 15'
print(f"-> Fila {fila_especifica} imputada correctamente con 'Laptop Pro 15'.")

# 2. Limpiar nulos o vacíos en el producto
# Creamos df_1_limpio a partir de aquí para trabajar sobre una nueva copia 
df_1_limpio = df_1[df_1['producto'].notna() & (df_1['producto'] != '')].copy()

# CONTAMOS: ¿Cuántos eran nulos realmente?
valores_nulos = len(df_1_limpio)
descartados_por_nulos = registros_iniciales - valores_nulos

# 3. Filtrar cantidades mayores a 0
df_1_limpio = df_1_limpio[df_1_limpio['cantidad'] > 0]

# CONTAMOS: ¿Cuántos se van por cantidad?
registros_finales = len(df_1_limpio)
descartados_por_cantidad = valores_nulos - registros_finales

# Total global
total_descartados = registros_iniciales - registros_finales

print("\n--- Reporte de Limpieza e Imputación ---")
print(f"Registros iniciales analizados:            {registros_iniciales}")
print(f"Eliminados por estar vacíos/nulos:         {descartados_por_nulos}")
print(f"Eliminados por cantidad <= 0:              {descartados_por_cantidad}")
print(f"Total de registros descartados:            {total_descartados}")
print(f"Registros finales:                         {registros_finales}")
```

```python
# Al observar se puede identificar que se han eliminado las filas con valores nulos incluyendo la fila que tenia un valor con cantidad < a 0
df_1_limpio.head(60)
```

#### **3.3 Reporte Mensual Generado*

```python
# Feature Engineering: Creación de nuevas columnas numéricas y temporales
df_1_limpio['ingreso_bruto'] = df_1_limpio['cantidad'] * df_1_limpio['precio_unitario']
df_1_limpio['ingreso_neto'] = df_1_limpio['ingreso_bruto'] * (1 - df_1_limpio['descuento'])

# Extraer el período mensual (Año-Mes) para la agrupación
df_1_limpio['mes'] = df_1_limpio['fecha'].dt.to_period('M')

# Agrupación mensual con agregaciones específicas (.groupby + .agg)
resumen_mensual = df_1_limpio.groupby('mes').agg(
    total_ingreso_neto=('ingreso_neto', 'sum'),
    total_ordenes=('order_id', 'count')
).reset_index()

print("\n--- Resumen Mensual Generado ---")
print(resumen_mensual)
```

```python
# Crear rango de meses
meses = pd.period_range(
    start=df_1_limpio['mes'].min(),
    end=df_1_limpio['mes'].max(),
    freq='M'
)

# Crear tabla de metas
df_metas = pd.DataFrame({
    'mes': meses,
    'meta_ingreso': [50000] * len(meses)
})

# Convertir ambas columnas 'mes' a texto para que el merge funcione
resumen_mensual['mes'] = resumen_mensual['mes'].astype(str)
df_metas['mes'] = df_metas['mes'].astype(str)

# Unir resumen mensual con metas
resumen_final = pd.merge(
    resumen_mensual,
    df_metas,
    on='mes',
    how='left'
)

# Calcular porcentaje de cumplimiento
resumen_final['porcentaje_cumplimiento'] = (
    resumen_final['total_ingreso_neto']
    / resumen_final['meta_ingreso']
) * 100

print("\n--- Resumen Final con Metas y Cumplimiento ---")
print(resumen_final)

# Crear conexión SQLite
engine = create_engine('sqlite:///techventas.db')

# Exportar a SQLite
resumen_final.to_sql(
    'ventas',
    con=engine,
    if_exists='replace',
    index=False
)

print("\n[OK] Datos exportados exitosamente a la tabla 'ventas'.")

# Verificación
query_verificacion = """
SELECT COUNT(*) AS total_registros
FROM ventas
"""

verificacion_df = pd.read_sql_query(
    query_verificacion,
    con=engine
)

print("\n--- Verificación desde SQLite ---")
print(
    f"Cantidad de registros en la tabla ventas: "
    f"{verificacion_df['total_registros'].iloc[0]}"
)
```
